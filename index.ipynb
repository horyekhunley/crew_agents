```python
import os
from dotenv import load_dotenv
from crewai_tools import SerperDevTool
from langchain_ollama import ChatOllama
from crewai import Agent, Task, Crew, Process
from IPython.display import Markdown, display
from langchain.tools.tavily_search import TavilySearchResults
from langchain.utilities.tavily_search import TavilySearchAPIWrapper

load_dotenv()

llm = ChatOllama(model="gemma2:2b", request_timeout=600, temperature=0)
```

```python
search_tool = SerperDevTool()

search = TavilySearchAPIWrapper()
tavily_tool = TavilySearchResults(api_wrapper=search)
```

```python
# Define your agents with roles and goals
researcher = Agent(
  role='Senior Research Analyst',
  goal='Uncover cutting-edge developments in AI and chemical engineering',
  backstory="""You work at a leading STEM think tank.
  Your expertise lies in identifying emerging trends in science, technology, engineering and mathematics.
  You have a knack for dissecting complex data and presenting actionable insights.""",
  verbose=True,
  allow_delegation=False,
  tools=[tavily_tool],
  llm=llm
)

writer = Agent(
  role='Tech Content research writer',
  goal='Craft compelling content on tech advancements in AI and chemical engineering',
  backstory="""You are a renowned Content Strategist, known for your insightful and engaging articles.
  You transform complex concepts into compelling narratives.""",
  verbose=True,
  allow_delegation=False,
  llm=llm
)
```

```python
# Create tasks for your agents
task1 = Task(
  description="""Conduct a comprehensive analysis of the latest advancements in AI and chemical engineering in the last decade.
  Identify key trends, breakthrough technologies, and potential industry impacts.""",
  expected_output="Full analysis report in bullet points",
  agent=researcher
)

task2 = Task(
  description="""Using the insights provided, develop an engaging blog
  post that highlights the most significant AI advancements.
  Your post should be informative yet accessible, catering to a tech-savvy audience.
  Make it sound cool, avoid complex words so it doesn't sound robotic or make it obviously sound like AI.""",
  expected_output="Full blog post of at least 4 paragraphs",
  agent=writer
)
```

```python
# Instantiate your crew with a sequential process
crew = Crew(
  agents=[researcher, writer],
  tasks=[task1, task2],
  verbose=True, # You can set it to 1 or 2 to different logging levels
  process=Process.sequential
)

# Get your crew to work!
result = crew.kickoff()
```

```python
display(Markdown(str(result)))
```
